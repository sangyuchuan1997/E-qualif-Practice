#モデルの制作

```python
# ここからダウンロードされたデータに対してモデルを作成し、検証用データに対して予測結果を出力して下さい。
from sklearn.impute import SimpleImputer
from sklearn.preprocessing import LabelEncoder, OneHotEncoder


# 前処理
def preprocess_titanic(df):
    # コピーを作成
    df = df.copy()
    
    # 全てのカテゴリ列の可能な値を事前定義
    POSSIBLE_TITLES = ['Mr', 'Miss', 'Mrs', 'Master', 'Rare']
    POSSIBLE_DECKS = ['A', 'B', 'C', 'D', 'E', 'F', 'G', 'T', 'U']
    POSSIBLE_EMBARKED = ['C', 'Q', 'S']

    # Age, Fareの欠損値を中央値で埋める
    imputer = SimpleImputer(strategy='median')
    df['age'] = imputer.fit_transform(df[['age']])
    df['fare'] = imputer.fit_transform(df[['fare']])

    # カテゴリ変数のエンコーディング
    le = LabelEncoder()
    df['sex'] = le.fit_transform(df['sex'])

    # Embarkedの欠損値を最頻値で埋める
    df['embarked'].fillna(df['embarked'].mode()[0], inplace=True)
    

    # 敬称を抽出
    df['Title'] = df['name'].str.extract('([A-Za-z]+)\.', expand=False)
    # 珍しい敬称をまとめる
    rare_titles = ['Lady', 'Countess', 'Capt', 'Col', 'Don',
                   'Dr', 'Major', 'Rev', 'Sir', 'Jonkheer', 'Dona']
    df['Title'] = df['Title'].replace(rare_titles, 'Rare')
    df['Title'] = df['Title'].replace('Mlle', 'Miss')
    df['Title'] = df['Title'].replace('Ms', 'Miss')
    df['Title'] = df['Title'].replace('Mme', 'Mrs')

    # Cabinからデッキ情報を抽出
    df['Deck'] = df['cabin'].str[0]
    df['Deck'].fillna('U', inplace=True)

    # 家族サイズを作成
    df['FamilySize'] = df['sibsp'] + df['parch'] + 1
    df['IsAlone'] = (df['FamilySize'] == 1).astype(int)

    # Embarkedのone-hotエンコーディング
    for val in POSSIBLE_EMBARKED:
        df[f'Embarked_{val}'] = (df['embarked'] == val).astype(int)

    # Titleのone-hotエンコーディング
    for val in POSSIBLE_TITLES:
        df[f'Title_{val}'] = (df['Title'] == val).astype(int)

    # Deckのone-hotエンコーディング
    for val in POSSIBLE_DECKS:
        df[f'Deck_{val}'] = (df['Deck'] == val).astype(int)

    # 不要な列を削除
    columns_to_drop = ['name', 'ticket', 'cabin', 'embarked', 'Title', 'Deck', 'boat', 'body', 'home.dest']
    df.drop(columns_to_drop, axis=1, inplace=True)

    return df


processed_data = preprocess_titanic(train_df)
processed_data.info()
```

```python
# モデルの構築
from sklearn.model_selection import GridSearchCV
from sklearn.ensemble import RandomForestClassifier

def build_model(X, y):
    # 基本モデル
    rf = RandomForestClassifier(random_state=42)
    
    # パラメータグリッド
    param_grid = {
        'n_estimators': [100, 300, 500],
        'max_depth': [10, 20, 30, None],
        'min_samples_split': [2, 5, 10],
        'min_samples_leaf': [1, 2, 4],
    }
    
    # グリッドサーチ
    grid_search = GridSearchCV(
        estimator=rf,
        param_grid=param_grid,
        cv=5,
        n_jobs=-1,
        scoring='accuracy',
    )
    
    grid_search.fit(X, y)
    
    # 最適パラメータと結果
    results = {
        'best_params': grid_search.best_params_,
        'best_score': grid_search.best_score_,
        'model': grid_search.best_estimator_,
    }
    
    # 特徴量の重要度
    feature_importances = grid_search.best_estimator_.feature_importances_
    results['feature_importances'] = feature_importances
    
    return results

X = processed_data.drop('survived', axis=1)
y = processed_data['survived']
results = build_model(X, y)
```

```python
# 結果の表示
print(f"Best accuracy: {results['best_score']:.4f}")
print("\nTop 10 important features:")
print(results['feature_importances'][:10])

# ベストモデル
model = results['model']
```
